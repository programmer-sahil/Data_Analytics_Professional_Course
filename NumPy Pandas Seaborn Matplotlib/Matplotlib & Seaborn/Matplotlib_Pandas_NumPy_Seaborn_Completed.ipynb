```python
# how to complete this project step by step?

'''
Project Flow:

1. Import libraries
2. Create dataset
3. View dataset
4. Basic analysis
5. Bar chart
6. Line chart
7. Pie chart

'''
```

```python
# Student Marks Analysis Dashboard
# Beginner Project using Pandas and Matplotlib

print("Student Marks Analysis Dashboard")
```

```python
# step 1: Import libraries

# import pandas for table data
import pandas as pd

# import matplotlib for charts
import matplotlib.pyplot as plt

print("Libraries imported successfully")
```

```python
# step 2: Create dataset

# create student marks dataset using dictionary

data = {
    "Name": ["Madhu", "Rahul", "Priya", "Arjun", "Neha"],
    "Math": [85, 78, 92, 74, 88],
    "Science": [90, 82, 89, 70, 91],
    "English": [88, 75, 95, 72, 84]
}

# convert dictionary into DataFrame
df = pd.DataFrame(data)

# view DataFrame
df
```

```python
# step 3: View basic dataset info

# show first rows
print("First 5 rows:")
print(df.head())


# show shape
print("\nShape of dataset:")
print(df.shape)


# show column names
print("\nColumns:")
print(df.columns)
```

```python
# step 4: Add total marks column

# create a new column Total
# add marks of all subjects

df["Total"] = df["Math"] + df["Science"] + df["English"]

df
```

```python
# step 5: Add average marks column

# create Average column
# divide total by 3 subjects

df["Average"] = df["Total"] / 3

df
```

```python
# step 6: Basic analysis

# find highest total marks
print("Highest Total Marks:", df["Total"].max())

# find lowest total marks
print("Lowest Total Marks:", df["Total"].min())

# find average of total marks
print("Average Total Marks:", df["Total"].mean())
```

```python
# step 7: Find topper

# find student with highest total marks
# Pandas library in Python, idxmax() is a method used to find the index label of the first occurrence of the maximum value in a Series or DataFrame

# idxmax() gives index of highest value
# loc[] selects that row

topper = df.loc[df["Total"].idxmax()]

print("Topper Details:")
print(topper)
```

```python
# step 8: Bar chart

# bar chart for total marks of students

plt.figure(figsize=(8,5))
plt.bar(df["Name"], df["Total"])
plt.title("Total Marks of Students")
plt.xlabel("Student Name")
plt.ylabel("Total Marks")
plt.show()
```

```python
# step 9: Line chart

# line chart for average marks of students

plt.figure(figsize=(8,5))
plt.plot(df["Name"], df["Average"], marker="o")
plt.title("Average Marks of Students")
plt.xlabel("Student Name")
plt.ylabel("Average Marks")
plt.show()

# line chart shows trend
# marker="o" shows points clearly
```

```python
# step 10: Pie chart

# pie chart for total marks distribution

plt.figure(figsize=(7,7))
plt.pie(df["Total"], labels=df["Name"], autopct="%1.1f%%")
plt.title("Marks Distribution of Students")
plt.show()

# pie chart shows percentage share
# autopct shows percentage on chart
```

```python
# step 11: Subject-wise average

# calculate average marks of each subject

subject_avg = {
    "Math": df["Math"].mean(),
    "Science": df["Science"].mean(),
    "English": df["English"].mean()
}

print(subject_avg)
```

```python
# step 12: Bar chart for subject averages

# bar chart of subject average marks

plt.figure(figsize=(8,5))
plt.bar(subject_avg.keys(), subject_avg.values())
plt.title("Average Marks by Subject")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.show()

# helps students compare subjects
# now chart is subject-wise, not student-wise
```

```python
# step 13: Grade column

# create grade based on average marks

grades = []

for avg in df["Average"]:
    if avg >= 90:
        grades.append("A")
    elif avg >= 80:
        grades.append("B")
    elif avg >= 70:
        grades.append("C")
    else:
        grades.append("D")

df["Grade"] = grades

df
```

```python
# step 14: Grade count

# count how many students got each grade

grade_count = df["Grade"].value_counts()

print(grade_count)
```

```python
# step 15: Pie chart for grade distribution

# pie chart for grade distribution

plt.figure(figsize=(7,7))
plt.pie(grade_count, labels=grade_count.index, autopct="%1.1f%%")
plt.title("Grade Distribution")
plt.show()
```

```python
# Pandas: value_counts()

# value_counts() counts repeated values
# very useful for categories like department, city, grade, gender

import pandas as pd

# Create simple student data
data = {
    "Name": ["Ali", "Sara", "John", "Ali", "Sara", "Ali"],
    "Department": ["CSE", "CSE", "ECE", "CSE", "ECE", "CSE"],
    "Marks": [85, 90, 78, 88, 91, 95]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Count how many times each name appears
print("\nCount of each student name:")
print(df["Name"].value_counts())

# Count how many students are in each department
print("\nCount of each department:")
print(df["Department"].value_counts())
```

```python
# Pandas: groupby()

# groupby() means make groups
# here we grouped students by department
# then found average marks of each group

import pandas as pd

data = {
    "Name": ["Ali", "Sara", "John", "David", "Emma"],
    "Department": ["CSE", "CSE", "ECE", "ECE", "CSE"],
    "Marks": [85, 90, 78, 88, 91]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Group by department and find average marks
grouped = df.groupby("Department")["Marks"].mean()

print("\nAverage marks department-wise:")
print(grouped)
```

```python
# Pandas: Remove duplicates

# duplicated() shows True for repeated rows
# drop_duplicates() removes repeated rows

import pandas as pd

data = {
    "Name": ["Ali", "Sara", "Ali", "John", "Sara"],
    "Marks": [85, 90, 85, 78, 90]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Check duplicate rows
print("\nDuplicate rows:")
print(df.duplicated())

# Remove duplicate rows
df_no_duplicates = df.drop_duplicates()

print("\nDataFrame after removing duplicates:")
print(df_no_duplicates)
```

```python
# Pandas: String operations

# .str.upper() → uppercase
# .str.title() → first letter capital
# .str.len() → length of text

import pandas as pd

data = {
    "Name": ["ali", "sara", "john", "david"],
    "City": ["kolkata", "delhi", "mumbai", "chennai"]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Convert names to uppercase
df["Name_Upper"] = df["Name"].str.upper()

# Convert city names to title case
df["City_Title"] = df["City"].str.title()

# Find length of each name
df["Name_Length"] = df["Name"].str.len()

print("\nAfter string operations:")
print(df)
```

```python
# Pandas: apply()

# apply() is used when built-in functions are not enough
# we created our own function and applied it to each mark

import pandas as pd

data = {
    "Name": ["Ali", "Sara", "John", "David"],
    "Marks": [85, 42, 78, 33]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Function to assign pass/fail
def check_result(mark):
    if mark >= 40:
        return "Pass"
    else:
        return "Fail"

# Apply function on Marks column
df["Result"] = df["Marks"].apply(check_result)

print("\nDataFrame after apply():")
print(df)
```

```python
# unique() and nunique()

import pandas as pd

data = {
    "City": ["Kolkata", "Delhi", "Mumbai", "Delhi", "Kolkata", "Chennai"]
}

df = pd.DataFrame(data)

print(df)

# Show unique values
print(df["City"].unique())

# Count number of unique values
print(df["City"].nunique())
```

```python
# map()

import pandas as pd

data = {
    "Name": ["Ali", "Sara", "John"],
    "Grade": ["A", "B", "C"]
}

df = pd.DataFrame(data)

print(df)

# Map grades to remarks
remarks = {
    "A": "Excellent",
    "B": "Good",
    "C": "Average"
}

df["Remark"] = df["Grade"].map(remarks)

print(df)
```

```python
# replace()

import pandas as pd

data = {
    "Name": ["Ali", "Sara", "John"],
    "Department": ["CS", "IT", "CS"]
}

df = pd.DataFrame(data)

print(df)

# Replace old values with new values
df["Department"] = df["Department"].replace("CS", "CSE")

print(df)
```

```python
# filtering with multiple conditions using & and |

import pandas as pd

data = {
    "Name": ["Ali", "Sara", "John", "David"],
    "Marks": [85, 35, 60, 28],
    "Department": ["CSE", "ECE", "CSE", "EEE"]
}

df = pd.DataFrame(data)

print(df)

# AND condition
print(df[(df["Marks"] > 40) & (df["Department"] == "CSE")])

# OR condition
print(df[(df["Marks"] > 80) | (df["Department"] == "EEE")])
```

```python
# proper use of loc[] and iloc[] (Row Select)

import pandas as pd

data = {
    "Name": ["Ali", "Sara", "John", "David"],
    "Marks": [85, 90, 78, 88],
    "Department": ["CSE", "ECE", "EEE", "CSE"]
}

df = pd.DataFrame(data)

print(df)

# loc[] uses label names
print(df.loc[0])                  # first row
print(df.loc[1:2, ["Name", "Marks"]])   # selected rows and columns

# iloc[] uses index positions
print(df.iloc[0])                 # first row
print(df.iloc[1:3, 0:2])          # rows 1 to 2, columns 0 to 1
```

```python
# aggregation with agg()

import pandas as pd

data = {
    "Department": ["CSE", "CSE", "ECE", "ECE", "CSE"],
    "Marks": [85, 90, 78, 88, 91]
}

df = pd.DataFrame(data)

print(df)

# Multiple aggregations
print(df.groupby("Department")["Marks"].agg(["mean", "max", "min", "sum"]))
```

```python
# sorting by multiple columns

import pandas as pd

data = {
    "Name": ["Ali", "Sara", "John", "David"],
    "Department": ["CSE", "CSE", "ECE", "ECE"],
    "Marks": [85, 90, 90, 78]
}

df = pd.DataFrame(data)

print(df)

# Sort by Department first, then Marks descending
sorted_df = df.sort_values(by=["Department", "Marks"], ascending=[True, False])

print(sorted_df)
```

```python
# set_index() and reset_index()

import pandas as pd

data = {
    "Roll": [101, 102, 103],
    "Name": ["Ali", "Sara", "John"],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)

print(df)

# Set Roll as index
df2 = df.set_index("Roll")
print(df2)

# Reset index back to normal
df3 = df2.reset_index()
print(df3)
```

```python
# datetime conversion with to_datetime()

import pandas as pd

data = {
    "Name": ["Ali", "Sara", "John"],
    "Join_Date": ["2024-01-15", "2024-02-10", "2024-03-20"]
}

df = pd.DataFrame(data)

print(df)

# Convert string date into datetime format
df["Join_Date"] = pd.to_datetime(df["Join_Date"])

print(df)
print(df.dtypes)
```

```python
# .dt accessor

import pandas as pd

data = {
    "Name": ["Ali", "Sara", "John"],
    "Join_Date": ["2024-01-15", "2024-02-10", "2024-03-20"]
}

df = pd.DataFrame(data)

df["Join_Date"] = pd.to_datetime(df["Join_Date"])

# Extract year, month, day
df["Year"] = df["Join_Date"].dt.year
df["Month"] = df["Join_Date"].dt.month
df["Day"] = df["Join_Date"].dt.day

print(df)
```

```python
# pivot table

import pandas as pd

data = {
    "Department": ["CSE", "CSE", "ECE", "ECE", "CSE"],
    "Gender": ["Male", "Female", "Male", "Female", "Male"],
    "Marks": [85, 90, 78, 88, 91]
}

df = pd.DataFrame(data)

print(df)

# Create pivot table
pivot = df.pivot_table(values="Marks", index="Department", columns="Gender", aggfunc="mean")

print(pivot)
```

```python
# crosstab()

import pandas as pd

data = {
    "Department": ["CSE", "CSE", "ECE", "ECE", "CSE"],
    "Gender": ["Male", "Female", "Male", "Female", "Male"]
}

df = pd.DataFrame(data)

print(df)

# Frequency table
table = pd.crosstab(df["Department"], df["Gender"])

print(table)
```

```python
# concat()

import pandas as pd

df1 = pd.DataFrame({
    "Name": ["Ali", "Sara"],
    "Marks": [85, 90]
})

df2 = pd.DataFrame({
    "Name": ["John", "David"],
    "Marks": [78, 88]
})

print(df1)
print(df2)

# Combine rows
result = pd.concat([df1, df2])

print(result)
```

```python
# merge()

import pandas as pd

students = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Ali", "Sara", "John"]
})

marks = pd.DataFrame({
    "ID": [1, 2, 3],
    "Marks": [85, 90, 78]
})

print(students)
print(marks)

# Merge using common column
result = pd.merge(students, marks, on="ID")

print(result)
```

```python
# join()

import pandas as pd

df1 = pd.DataFrame({
    "Name": ["Ali", "Sara", "John"]
}, index=[1, 2, 3])

df2 = pd.DataFrame({
    "Marks": [85, 90, 78]
}, index=[1, 2, 3])

print(df1)
print(df2)

# Join based on index
result = df1.join(df2)

print(result)
```

```python
# basic EDA workflow

import pandas as pd

data = {
    "Name": ["Ali", "Sara", "John", "David", "Sara"],
    "Department": ["CSE", "CSE", "ECE", "EEE", "CSE"],
    "Marks": [85, 90, None, 78, 90]
}

df = pd.DataFrame(data)

# 1. View first rows
print(df.head())

# 2. Check data types and missing values
print(df.info())

# 3. Statistical summary
print(df.describe())

# 4. Check missing values
print(df.isnull().sum())

# 5. Fill missing values
df["Marks"] = df["Marks"].fillna(df["Marks"].mean())

# 6. Check duplicate rows
print(df.duplicated().sum())

# 7. Remove duplicates
df = df.drop_duplicates()

# 8. Count departments
print(df["Department"].value_counts())

# 9. Group by department
print(df.groupby("Department")["Marks"].mean())

# 10. Final cleaned data
print(df)
```

```python
# NumPy introduction

# NumPy is used for fast numerical calculations
# NumPy array is similar to a Python list, but more powerful

import numpy as np

# Create a NumPy array
arr = np.array([10, 20, 30, 40, 50])

print("NumPy Array:")
print(arr)

print("\nType of arr:")
print(type(arr))
```

```python
# 1D and 2D arrays

# 1D array = one row of values
# 2D array = rows and columns like table

import numpy as np

# 1D array
arr1 = np.array([1, 2, 3, 4, 5])

# 2D array (Row then Column)
arr2 = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print("1D Array:")
print(arr1)

print("\n2D Array:")
print(arr2)
```

```python
# Shape and dimension

# shape tells rows and columns
# ndim tells dimension

import numpy as np

arr1 = np.array([10, 20, 30, 40])
arr2 = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print("Shape of arr1:", arr1.shape)
print("Dimension of arr1:", arr1.ndim)

print("Shape of arr2:", arr2.shape)
print("Dimension of arr2:", arr2.ndim)
```

```python
# Indexing in NumPy

import numpy as np

arr = np.array([10, 20, 30, 40, 50])

print("Full array:")
print(arr)

print("\nFirst element:", arr[0])
print("Third element:", arr[2])
print("Last element:", arr[-1])
```

```python
# Slicing in NumPy

import numpy as np

arr = np.array([10, 20, 30, 40, 50, 60])

print("Original array:")
print(arr)

print("\nElements from index 1 to 4:")
print(arr[1:5])

print("\nFirst 3 elements:")
print(arr[:3])

print("\nLast 2 elements:")
print(arr[-2:])
```

```python
# 2D array indexing

# first value is row index
# second value is column index

import numpy as np

arr = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

print("2D Array:")
print(arr)

print("\nElement at row 0, column 1:", arr[0, 1])
print("Element at row 1, column 2:", arr[1, 2])
```

```python
# Mathematical operations

# NumPy can do element-wise operations very easily

import numpy as np

arr1 = np.array([10, 20, 30])
arr2 = np.array([1, 2, 3])

print("Array 1:", arr1)
print("Array 2:", arr2)

print("\nAddition:")
print(arr1 + arr2)

print("\nSubtraction:")
print(arr1 - arr2)

print("\nMultiplication:")
print(arr1 * arr2)

print("\nDivision:")
print(arr1 / arr2)
```

```python
# Useful NumPy functions

import numpy as np

arr = np.array([10, 20, 30, 40, 50])

print("Array:", arr)

print("\nSum:", np.sum(arr))
print("Mean:", np.mean(arr))
print("Maximum:", np.max(arr))
print("Minimum:", np.min(arr))
```

```python
# Special arrays

import numpy as np

# Array of zeros
a = np.zeros(5)

# Array of ones
b = np.ones(5)

# Range of values
c = np.arange(1, 11)

print("Zeros array:")
print(a)

print("\nOnes array:")
print(b)

print("\nRange array:")
print(c)
```

```python
# shape, ndim, size

import numpy as np

# Create a 2D array
arr = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

print("Array:")
print(arr)

# shape = number of rows and columns
print("\nShape:", arr.shape)

# ndim = number of dimensions
print("Number of dimensions:", arr.ndim)

# size = total number of elements
print("Total number of elements:", arr.size)
```

```python
# Data Types with dtype and type conversion

import numpy as np

# Create array of integers
arr = np.array([1, 2, 3, 4, 5])

print("Original array:")
print(arr)

# Check data type
print("\nData type:", arr.dtype)

# Convert integer array to float array
arr_float = arr.astype(float)

print("\nConverted to float:")
print(arr_float)
print("New data type:", arr_float.dtype)
```

```python
# zeros(), ones(), arange(), linspace()

import numpy as np

# Array of zeros
a = np.zeros(5)
print("zeros():")
print(a)

# Array of ones
b = np.ones(5)
print("\nones():")
print(b)

# Range of values
c = np.arange(1, 11)
print("\narange():")
print(c)

# Evenly spaced values
d = np.linspace(0, 10, 5)
print("\nlinspace():")
print(d)
```

```python
# reshaping with reshape()

import numpy as np

# Create 1D array
arr = np.array([1, 2, 3, 4, 5, 6])

print("Original array:")
print(arr)

# Reshape into 2 rows and 3 columns
reshaped = arr.reshape(2, 3)

print("\nReshaped array (2 x 3):")
print(reshaped)
```

```python
# flattening with flatten() / ravel()

import numpy as np

# Create 2D array
arr = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print("Original 2D array:")
print(arr)

# Convert into 1D array using flatten()
flat1 = arr.flatten()
print("\nFlatten using flatten():")
print(flat1)

# Convert into 1D array using ravel()
flat2 = arr.ravel()
print("\nFlatten using ravel():")
print(flat2)
```

```python
# boolean masking

import numpy as np

arr = np.array([10, 20, 30, 40, 50])

print("Original array:")
print(arr)

# Create boolean mask
mask = arr > 25

print("\nBoolean mask (arr > 25):")
print(mask)

# Use mask to get values
print("\nValues greater than 25:")
print(arr[mask])
```

```python
# conditional filtering

import numpy as np

arr = np.array([5, 12, 18, 25, 30, 7])

print("Original array:")
print(arr)

# Filter values greater than 15
filtered = arr[arr > 15]

print("\nFiltered values greater than 15:")
print(filtered)
```

```python
# broadcasting


import numpy as np

arr = np.array([10, 20, 30, 40])

print("Original array:")
print(arr)

# Add one number to all elements
result = arr + 5

print("\nAfter adding 5 to all elements:")
print(result)

# Multiply all elements by 2
result2 = arr * 2

print("\nAfter multiplying by 2:")
print(result2)
```

```python
# axis concept

import numpy as np

arr = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

print("Array:")
print(arr)

# Sum column-wise
print("\nSum with axis=0 (column-wise):")
print(np.sum(arr, axis=0))

# Sum row-wise
print("\nSum with axis=1 (row-wise):")
print(np.sum(arr, axis=1))
```

```python
# statistical functions: mean(), median(), std(), var()

import numpy as np

arr = np.array([10, 20, 30, 40, 50])

print("Array:")
print(arr)

print("\nMean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))
print("Variance:", np.var(arr))
```

```python
# min(), max(), sum(), cumsum()


import numpy as np

arr = np.array([10, 20, 30, 40, 50])

print("Array:")
print(arr)

print("\nMinimum:", np.min(arr))
print("Maximum:", np.max(arr))
print("Sum:", np.sum(arr))
print("Cumulative Sum:", np.cumsum(arr))
```

```python
# sorting with sort()

import numpy as np

arr = np.array([40, 10, 30, 20, 50])

print("Original array:")
print(arr)

# Sort array
sorted_arr = np.sort(arr)

print("\nSorted array:")
print(sorted_arr)
```

```python
# copying vs viewing

import numpy as np

arr = np.array([10, 20, 30, 40])

print("Original array:")
print(arr)

# View shares memory
view_arr = arr.view()

# Copy creates separate memory
copy_arr = arr.copy()

# Change original array
arr[0] = 999

print("\nAfter changing original array:")
print("Original array:", arr)
print("View array:", view_arr)
print("Copy array:", copy_arr)
```

```python
# random module

import numpy as np

# Random integer values
rand_int = np.random.randint(1, 10, size=5)
print("Random integers:")
print(rand_int)

# Random float values between 0 and 1
rand_float = np.random.rand(5)
print("\nRandom floats:")
print(rand_float)

# Random 2D array
rand_2d = np.random.randint(1, 100, size=(2, 3))
print("\nRandom 2D array:")
print(rand_2d)
```

```python
# stacking arrays

import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("Array a:")
print(a)

print("\nArray b:")
print(b)

# Horizontal stacking
h_stack = np.hstack((a, b))
print("\nHorizontal stack:")
print(h_stack)

# Vertical stacking
v_stack = np.vstack((a, b))
print("\nVertical stack:")
print(v_stack)
```

```python
# splitting arrays

import numpy as np

arr = np.array([10, 20, 30, 40, 50, 60])

print("Original array:")
print(arr)

# Split into 3 equal parts
split_arr = np.split(arr, 3)

print("\nSplit array into 3 parts:")
print(split_arr)
```

```python
# universal functions (sqrt, log, exp, etc.)


import numpy as np

arr = np.array([1, 4, 9, 16])

print("Array:")
print(arr)

print("\nSquare root:")
print(np.sqrt(arr))

print("\nExponential:")
print(np.exp(arr))

print("\nLogarithm:")
print(np.log(arr))
```

```python
# handling missing values basics (nan)

import numpy as np

arr = np.array([10, 20, np.nan, 40, 50])

print("Array with missing value:")
print(arr)

# Check missing values
print("\nIs NaN present?")
print(np.isnan(arr))

# Mean ignoring NaN
print("\nMean ignoring NaN:")
print(np.nanmean(arr))

# Sum ignoring NaN
print("\nSum ignoring NaN:")
print(np.nansum(arr))
```

```python
# basic linear algebra functions

import numpy as np

a = np.array([
    [1, 2],
    [3, 4]
])

b = np.array([
    [5, 6],
    [7, 8]
])

print("Matrix A:")
print(a)

print("\nMatrix B:")
print(b)

# Matrix multiplication
print("\nMatrix multiplication:")
print(np.dot(a, b))

# Transpose
print("\nTranspose of A:")
print(a.T)

# Determinant
print("\nDeterminant of A:")
print(np.linalg.det(a))
```

```python
# practical NumPy for data analytics workflow

import numpy as np

# Sales data of 7 days
sales = np.array([200, 250, 180, 300, 280, 220, 260])

print("Sales data:")
print(sales)

# Basic analysis
print("\nTotal sales:", np.sum(sales))
print("Average sales:", np.mean(sales))
print("Maximum sales:", np.max(sales))
print("Minimum sales:", np.min(sales))

# Days where sales are greater than 240
print("\nSales greater than 240:")
print(sales[sales > 240])

# Add bonus 20 to all days
updated_sales = sales + 20
print("\nUpdated sales after bonus:")
print(updated_sales)

# Sort sales
print("\nSorted sales:")
print(np.sort(sales))
```

```python
# Program 3: Create a simple line plot

import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")

# Line plot between total_bill and tip
sns.lineplot(x="total_bill", y="tip", data=df)

plt.title("Line Plot of Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
# Program 4: Create a bar plot

import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")

# Bar plot showing average total bill by day
sns.barplot(x="day", y="total_bill", data=df)

plt.title("Average Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Average Total Bill")
plt.show()
```

```python
# Program 5: Create a count plot

import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")

# Count how many records are present for each day
sns.countplot(x="day", data=df)

plt.title("Count of Records by Day")
plt.xlabel("Day")
plt.ylabel("Count")
plt.show()
```

```python
# Program 6: Create a scatter plot

import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")

# Scatter plot between total bill and tip
sns.scatterplot(x="total_bill", y="tip", data=df)

plt.title("Scatter Plot of Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
# Program 7: Create a histogram

import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")

# Histogram of total_bill
sns.histplot(df["total_bill"], bins=10, kde=False)

plt.title("Histogram of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

```python
# Program 8: Histogram with KDE curve

import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")

# Histogram with KDE
sns.histplot(df["total_bill"], bins=10, kde=True)

plt.title("Histogram with KDE of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

```python
# Program 9: Create a box plot

import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")

# Box plot of total bill by day
sns.boxplot(x="day", y="total_bill", data=df)

plt.title("Box Plot of Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
# Program 10: Create a violin plot

import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")

# Violin plot of total bill by day
sns.violinplot(x="day", y="total_bill", data=df)

plt.title("Violin Plot of Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
# Program 11: Create a pair plot

import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")

# Pair plot for numeric columns
sns.pairplot(df)

plt.show()
```

```python
# Program 12: Create a heatmap using correlation matrix

import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")

# Select only numeric columns
numeric_df = df.select_dtypes(include=["number"])

# Create correlation matrix
corr = numeric_df.corr()

# Display heatmap
sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```

```python
# Program 13: Use hue to compare categories

import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")

# Scatter plot with hue
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=df)

plt.title("Total Bill vs Tip by Gender")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
# Program 14: Change Seaborn style

import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn theme
sns.set_theme(style="darkgrid")

df = sns.load_dataset("tips")

sns.barplot(x="day", y="tip", data=df)

plt.title("Average Tip by Day")
plt.show()
```
